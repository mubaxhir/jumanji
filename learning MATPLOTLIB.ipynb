# Error bar chart

```python
import matplotlib.pyplot as ply
import numpy as np
x = np.arange(0, 4, 0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x, y,yerr=e1, fmt='.-');  # for error bar chart
plt.show()
```

in above graph we just plotted x versus y with error deltas as vertical error bars as specified by the yerr keyword argument.
there is an equivalent argument,xeer,to draw horizontal error bars

fmt: this is the plot format for lines . if none then only the error bars are plotted with no line connecting them.

ecolor:this accepts any Matplotlib color and specifies the color of the error bar . if none the marker color is used.

elinewidth:this specifies the size of the crap of the error bar in pixels

```python
e2 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x, y, yerr= e1, xerr=e2, fmt = '.-',capsize=0) # each point gives four direction errors -y,+y,-x,+x
plt.show()
```

```python
plt.errorbar(x, y, yerr=[ e1,e2], fmt = '.-',capsize=0,ecolor='red') #-y is e1   &  +y is e2
plt.show()
```


# Bar charts

```python
import matplotlib.pyplot as plt
plt.bar([1,2,3],[3,2,5]);
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
dict1 = {'a':40,'b':70,'c':30,'d':85}
for i, key in enumerate(dict1):plt.bar(i,dict1[key]);
plt.xticks(np.arange(len(dict1))+0.4,dict1.keys());
plt.yticks(dict1.values());
plt.show()
```

```python
data1 = 10*np.random.rand(5)
data2 = 10*np.random.rand(5)
data3 = 10*np.random.rand(5)
e2 = 0.5 * np.abs(np.random.randn(len(data2)))
locs = np.arange(1,len(data1)+1)
width = 0.27
plt.bar(locs,data1,width=width,color='yellow');
plt.bar(locs+width,data2,yerr=e2,width=width,color='red'); 
plt.bar(locs+2*width,data3,width=width,color='green');
plt.xticks(locs + width*1.5 + locs);
plt.savefig
plt.show()
```
